v. Explorasi Data

```python
## Proses query
df[(df['Def'] > 68) & # mengambil data pokemon yang memiliki def lebih besar dari 68
     (df['Type'] == 'Grass') & # mengambil data pokemon bertype Grass
     (df['Speed'] > 70) & # mengambil data pokemon yang mempunya kecepatan lebih dari 70
     (df['Sp. Atk'] >= 65)].sort_values(by='Sp. Atk', ascending = True) # sp. atk lebih besar sama dengan 65
```

Tabel diatas merupakan hasil dari pencarian pokemon yang memiliki Def lebih besar dari 68, ber type Grass, Mempunyai Speed atau Kecepatan lebih besar dari 70, Dan Special Attack lebih besar sama dengan 65, setelah data ditemukan diurutkan berdasarkan Special Attack dari nilai terkecil hingga terbesar
