v. Explorasi Data

```python
## Proses query
df[(df['Legendary'] == True) & 
   (df['Type'] == 'Grass') & 
   (df['Speed'] > 70)].sort_values(by='Speed', ascending = True)
```

df[(df['Legendary'] == True) untuk menampilkan kolom Legendary dengan nilai True, (df['Type'] == 'Grass') menampilkan Pokemon type Grass saja, (df['Speed'] > 70) Menampilkan kolom speed yang bernilai lebih dari 70, sort_values(by='Speed', ascending = True) untuk mengurutkan nilai yang telah di dapat dari hasil pencarian berdasarkan Speed Terendah menuju Tertinggi, Symbol & ialah penghubung antara pencarian satu dengan yang lainnya

Tabel diatas merupakan hasil dari pencarian pokemon yang memiliki value Legendary, ber type Grass, Mempunyai Speed atau Kecepatan lebih besar dari 70, setelah data ditemukan diurutkan berdasarkan Speed dari nilai terkecil hingga terbesar

```python
df[(df['Legendary'] == True) & 
   (df['Type'] == 'Steel') & 
   (df['HP'] > 40)].sort_values(by='HP', ascending = True)
```

Disini Saya coba menampilkan pokemon legendary bertipe Steel atau baja, dan akan mencoba mensimulasikan pertarungan antar registeel dengan virizion yang bertipe grass, pokemon registeel mempunyai speed 50 sedangkan virizion mempunyai speed 108 (dapat dilihat pada tabel sebelumnya), saya asumsikan 2 kali pokemon menyerang putaran selanjutnya dapat mengeluarkan special atk

ketika virizion pertama menyerang , registeel kehilangan darah 20 dikarenakan def yang begitu tinggi dari registeel, lalu registeel balik menyerang, virizion terkena serangan 33, perlu diketahui pokemon bertipe grass lemah terhadap pokemon bertipe flying,poison,steel,dragon,bug dan fire

Posisi hp registeel sekarang 60, dan virizion 57, serangan ke 2 virizion berjumlah 24, registeel menyerang balik 29, hp registeel tersisa 36 sedangkan virizion 28, virizion bersiap melakukan serangan spesial terhadap registeel yang mempunyai kekebalan terhadap pokemon tipe grass, serangan spesial tersebut berhasil melukai registeel sebanyak 26, hp registeel sekarang tersisa 10, dan registeel yang bertipe lebih kuat ketika menghadapi pokemon tipe grass meluncurkan serangan spesial dan mengenai virizion sebanyak 27, virizion berhasil menahan serangan dari registeel dikarenakan mempunyai Spesial Defense yang tinggi yaitu 129, Lalu virizion pun menghabisi registeel dengan serangan normal nya, dan tersisa darah virizion 1 hp

```python
df[(df['Sp. Atk'] >= 0) & 
   (df['Type'] == 'Dragon') & 
   (df['Def'] <= 599)].sort_values(by='Def', ascending = True)
```

Menampilkan Tabel Pokemon Bertipe Dragon yang mempunyai Sp. Atk lebih dari sama dengan 0 , Def lkurang dari sama dengan 599
dan diurutkan berdasarkan Total nilai tertinggi ke terendah

```python
## Grouping dan Aggregating
df.groupby(by=['Type','Total'])[['HP']].mean().sort_values(by='Total', ascending = True)
```

Pokemon Bertype Dragon dan Psychic sama sama memiliki Pokemon nilai Total tertinggi yaitu 780, sedangkan kolom HP adalah Rata rata HP dari pokemon yang memiliki nilai Total Tersebut

Kolom Total diurutkan dari kecil Terbesar sehingga nilai Total Terkecil dimiliki oleh Pokemon Bertipe Grass selanjutnya Total nilai terkecil dimiliki oleh Pokemon bertipe Normal dan seterusnya

```python
df.groupby(by=['Type','Def'])[['Atk']].max()
```

Menentukan nilai terbesar dari kolom Atk , dengan Type dan Def sebagai pembanding nya

```python
df.groupby(by=['Gen','Legendary'])[['Sp. Def']].median()
```

df.groupby(by=['Gen','Legendary']) ialah proses grouping untuk menampilkan isi kolom Gen Di pisahkan kembali kedalam pokemon Legendary Atau Bukan
[['Sp. Def']].median() <- adalah proses untuk menentukan kolom yang mau di tampilkan berdasarkan nilai Tengah
Menghitung nilai Tengah kolom Sp. Def Dari setiap Gen Berdasarkan Legendary
